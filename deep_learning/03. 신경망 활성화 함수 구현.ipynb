# 계단 함수 구현

```python
def step_func(x_list) :
    y_list = list()
    for x in x_list : 
        if x > 0 :
            y_list.append(1) 
        else :
            y_list.append(0)
    return y_list
```

```python
import numpy as np 
X = np.arange(-3,3,0.001)
x = list(X)
y = step_func(x)

import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
```

# 넘파이 트릭으로 계단함수 구현하기

```python
def step_func_with_numpy_trick(x) :
    y = x > 0
    return y.astype(np.int)

# y.astype(np.int) : astype으로 y의 type을 int로 바꾸어줌  
```

```python
X = np.arange(-3,3,0.001)
y1 = step_func_with_numpy_trick(X)

plt.plot(X,y1)
plt.show()
```

# Sigmoid 함수 구현

```python
def Sigmoid (x) :
    return 1/(1+np.exp(-x))
```

```python
X = np.arange(-3,3,0.001)
y2 = Sigmoid(X)
plt.plot(X,y1,linestyle ="--", label="step")
plt.plot(X,y2, label = "sigmoid")
plt.ylim(-0.1, 1.1)
plt.show()
```

# Relu 함수 구현

```python
def relu(x) :
    return np.maximum(0,x)
```

```python
X = np.arange(-1,1,0.001)
y3 = relu(X)

plt.plot(X,y3, label = "relu")
plt.ylim(-0.1, 1.1)
plt.show()
```
